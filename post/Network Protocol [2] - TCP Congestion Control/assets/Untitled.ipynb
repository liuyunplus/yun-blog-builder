```python
import matplotlib.pyplot as plt
import numpy as np

def additive_increase():
    x = np.arange(0, 5, 1)
    y = 3*x
    return x, y

def binary_increase():
    x = np.arange(5, 20, 1)
    y = np.log(x) + 6
    return x, y

def slow_start():
    x = np.arange(21, 25, 1)
    y = x + 10
    return x, y

plt.ylim(0, 50)
plt.xlim(0, 30)
plt.plot(*additive_increase(), label='Binary Increase')
plt.plot(*binary_increase(), label='Additive Increase')
plt.plot(*slow_start(), label='Slow Start')
plt.xlabel('Time')
plt.ylabel('Window Size')
plt.legend()
plt.title('TCP BIC Window Growth')
plt.show()
```
